## Visualizations

```python
# 'foreign_gross'column contains values I want to convert to 'int64' for plotting purposes

# Remove commas from the column
df2['foreign_gross'] = df2['foreign_gross'].str.replace(',', '')

# Convert the column to float data type
df2['foreign_gross'] = df2['foreign_gross'].astype("float")

# Convert the column to integer data type
df2['foreign_gross'] = df2['foreign_gross'].astype("int64")
```

```python
#confirming the data type changed to 'int64'
df2['foreign_gross'].dtypes
```

```python
fig, ax = plt.subplots(figsize=(15,5))

# Plot box plots
ax.boxplot([df2['domestic_gross'], df2['foreign_gross']])

# Set x-axis tick labels
ax.set_xticklabels(['Domestic Gross', 'Foreign Gross'])

# Set y-axis label
ax.set_ylabel('Amount')

# Set title
plt.title('Box Plot of Box Office Performance')

# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))

# Plot histograms
ax.hist(df2['domestic_gross'], bins=10, label='Domestic Gross')
ax.hist(df2['foreign_gross'], bins=10, label='Foreign Gross')

# Set x-axis label
ax.set_xlabel('Amount')

# Set y-axis label
ax.set_ylabel('Frequency')

# Set title
plt.title('Histograms of Box Office Performance')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
