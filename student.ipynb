## Visualizations

```python
#creating a lineplot of the 'genre' column from df(first dataframe)
fig, ax = plt.subplots(figsize=(15,5))
genre_counts = df['genres'].value_counts()

genre_counts.plot()

plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Movie Genres')

plt.show()
```

```python
#creating a scatter plot of the 'averagerating' column and 'numvotes' column from df1(the second dataframe)
fig, ax = plt.subplots()

# Plot the scatter plot
ax.scatter(df1['averagerating'], df1['numvotes'])

# Set x-axis label
ax.set_xlabel('Rating')

# Set y-axis label
ax.set_ylabel('Number of Votes')

# Set title
plt.title('Movie Ratings vs. Number of Votes')

# Show the plot
plt.show();
```

```python
#creating a boxplot of the 'domestic_gross' column and 'foreign_gross' column from df2(the third dataframe)
fig, ax = plt.subplots(figsize=(15,5))

# Plot box plots
ax.boxplot([df2['domestic_gross'], df2['foreign_gross']])

# Set x-axis tick labels
ax.set_xticklabels(['Domestic Gross', 'Foreign Gross'])

# Set y-axis label
ax.set_ylabel('Amount')

# Set title
plt.title('Box Plot of Domestic and Foreign Gross')

# Show the plot
plt.show();
```

```python
#creating a histogram of the 'domestic_gross' column and 'foreign_gross' column from df2(the third dataframe)
fig, ax = plt.subplots(figsize=(15,5))

# Plot histograms
ax.hist(df2['domestic_gross'], bins=10, label='Domestic Gross')
ax.hist(df2['foreign_gross'], bins=10, label='Foreign Gross')

# Set x-axis label
ax.set_xlabel('Amount')

# Set y-axis label
ax.set_ylabel('Frequency')

# Set title
plt.title('Histograms of Domestic and Foreign Gross')

# Add a legend
ax.legend()

# Show the plot
plt.show();
```

```python
#creating a boxplot of the 'production_budget' column, 'domestic_gross' column and 'worldwide_gross' column 
#from df3(the fourth dataframe)
fig, ax = plt.subplots(figsize=(15,5))

# Plot box plots for the variables of interest
ax.boxplot([df3['domestic_gross'], df3['worldwide_gross'], df3['production_budget']])

# Set x-axis tick labels
ax.set_xticklabels(['Domestic Gross', 'Worldwide Gross', 'Production Budget'])

# Set y-axis label
ax.set_ylabel('Amount in USD')

# Set title
plt.title('Box Plot of Domestic Gross, Worldwide Gross and Production Budget')

plt.show();
```

```python
#creating a histogram of the 'production_budget' column, 'domestic_gross' column and 'worldwide_gross' column 
#from df3(the fourth dataframe)
fig, ax = plt.subplots(figsize=(15,5))

# Plot histograms for the variables of interest
ax.hist(df3['domestic_gross'], bins=10, alpha=0.5, label='Domestic Gross')
ax.hist(df3['worldwide_gross'], bins=10, alpha=0.5, label='Worldwide Gross')
ax.hist(df3['production_budget'], bins=10, alpha=0.5, label='Production Budget')

# Set x-axis label
ax.set_xlabel('Amount in USD')

# Set y-axis label
ax.set_ylabel('Frequency')

# Set title
plt.title('Histograms of Domestic Gross, Worldwide Gross and Production Budget')

# Add a legend
ax.legend()

# Show the plot
plt.show();
```
